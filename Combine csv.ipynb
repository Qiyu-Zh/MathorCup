```python
df = pd.read_csv(df_path)
df
```

```python
df = pd.merge(df, train_df[['carid', 'predPrice']], how='left')
df['predPrice'] = np.exp(df['predPrice'].values) - 1
df
```

```python
df['withdrawDate'].fillna(0, inplace=True)
df['withdrawDate'][df['withdrawDate'] != 0] = 1
df.rename(columns={'withdrawDate': 'isTraded'}, inplace=True)
df
```

```python
df['isCut'] = 0
df['isCut'][df['updatePriceTimeJson'] != '{}'] = 1
df
```

```python
df['cutNum'] = np.array([len(eval(df['updatePriceTimeJson'][i])) for i in range(len(df['updatePriceTimeJson']))])
df
```

```python
track_num = 3

cut_info = {'cutTime' + str(i + 1): [] for i in range(track_num)}
for i in range(track_num):
    cut_info['cutPrice' + str(i + 1)] = []

for i in range(len(df['updatePriceTimeJson'])):
    dic = eval(df['updatePriceTimeJson'][i])
    if len(dic) > track_num:
        for j in range(track_num):
            index = (j + 1) * len(dic) // track_num - 1
            cut_info['cutTime' + str(j + 1)].append(list(dic.keys())[index])
            cut_info['cutPrice' + str(j + 1)].append(list(dic.values())[index])
    else:
        for j in range(len(dic)):
            cut_info['cutTime' + str(j + 1)].append(list(dic.keys())[j])
            cut_info['cutPrice' + str(j + 1)].append(list(dic.values())[j])
        for j in range(len(dic), track_num):
            cut_info['cutTime' + str(j + 1)].append(np.NAN)
            cut_info['cutPrice' + str(j + 1)].append(np.NAN)
cut_df = pd.DataFrame(cut_info)
cut_df
```

```python
df = pd.concat([df, cut_df], axis=1)
df.drop(columns='updatePriceTimeJson', inplace=True)
```

```python
df
```

```python
df['dFirstCutTime'] = (pd.to_datetime(df['cutTime1']) - pd.to_datetime(df['pushDate'])).map(lambda x: x.days)
df['dSecondCutTime'] = (pd.to_datetime(df['cutTime2']) - pd.to_datetime(df['cutTime1'])).map(lambda x: x.days)
df['dThirdCutTime'] = (pd.to_datetime(df['cutTime3']) - pd.to_datetime(df['cutTime2'])).map(lambda x: x.days)
df['dCutTime'] = (pd.to_datetime(df['pullDate']) - pd.to_datetime(df['pushDate'])).map(lambda x: x.days)
df
```

```python
df.drop(columns=['pushDate', 'pullDate', 'cutTime1', 'cutTime2', 'cutTime3'], inplace=True)
```

```python
to_fp = ['cutPrice1', 'cutPrice2', 'cutPrice3']
for feature in to_fp:
    df[feature] = df[feature].astype(np.float)
```

```python
df['dFirstCutPrice'] = df['pushPrice'] - df['cutPrice1']
df['dSecondCutPrice'] = df['cutPrice1'] - df['cutPrice2']
df['dThirdCutPrice'] = df['cutPrice2'] - df['cutPrice3']
df['dCutPrice'] = [df['pushPrice'][i] - min(df['cutPrice1'][i], df['cutPrice2'][i], df['cutPrice3'][i]) for i in range(len(df))]
```

```python
df['dTruePrice'] = df['pushPrice'] - df['predPrice']
df['pTruePrice'] = df['dTruePrice'] / df['predPrice']

df['pFristCutPrice'] = df['dFirstCutPrice'] / df['pushPrice']
df['pSecondCutPrice'] = df['dSecondCutPrice'] / df['cutPrice1']
df['pThirdCutPrice'] = df['dThirdCutPrice'] / df['cutPrice2']
df['pCutPrice'] = df['dCutPrice'] / df['pushPrice']

df['pFirstAllTime'] = df['dFirstCutTime'] / df['dCutTime']
df['pFirstAllTime'].fillna(1, inplace=True)

df['pSecondAllTime'] = df['dSecondCutTime'] / df['dCutTime']
df['pSecondAllTime'].fillna(1, inplace=True)
```
